```python
def checksame(f, g): 
    f_contents = open(dir_name+f).read()
    g_contents = open(dir_name+g).read()
    return f_contents == g_contents
```

```python
def loadset(fn):
    stuff = open(dir_name+fn).read()
    return set(l.strip() for l in stuff.split('\n') if len(l) > 0)
def set_diff(f, g):
    fset = loadset(f)
    gset = loadset(g)
    return (fset-gset, gset-fset)
```

```python
def get_und_num(experiment_name, fn): 
    assert fn.startswith(experiment_name) and fn.endswith("_undecided.txt")
    rest = fn[len(experiment_name): -1 * len("_undecided.txt")]
    if rest == "all": 
        return None
    return int(rest)
```

```python
[(fn, get_und_num(cct_name, fn)) for fn in cct_undecideds]
```

```python
geq_200_cct_fns = [fn for fn in cct_undecideds if get_und_num(cct_name, fn) is not None and get_und_num(cct_name, fn) > 200]
sorted(geq_200_cct_fns)
```
